```python
# finding the ratio of males and females
m_students = len(df[df["sex"]=='M'])
f_students = len(df[df["sex"]=='F'])

print(" no. of male students : ", m_students)
print(" no. of female students : ", f_students)
```

```python
# find the grade average of all the grades for each student
df['gradeavg'] = (df['G1']+df['G2']+ df['G3'])/3
df['gradeavg']
```

```python
df['sex']=df['sex'].replace({'F':1, 'M':0})
df
```

```python
df.nunique()
```

```python
# creating a grading system

def define_grade(df):
    grades = []
    for row in df['gradeavg']:
        if row >= (0.9 * df['gradeavg'].max()):
            grades.append('A+')
        elif row >= (0.7 * df['gradeavg'].max()):
            grades.append('A')
        elif row < (0.7 * df['gradeavg'].max()):
            grades.append('B')
    df['grades'] = grades
    return df
```

```python
# displaying the grade using define
df = define_grade(df)
df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.utils import shuffle

df = df[["G1", "G2", "G3", "studytime", "failures", "absences"]]
predict = "G3"

x = np.array(df.drop([predict], 1))
y = np.array(df[predict])

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)


linear_regression = LinearRegression()
linear_regression.fit(xtrain, ytrain)
```

```python
accuracy = linear_regression.score(xtest, ytest)
print(accuracy)
```
